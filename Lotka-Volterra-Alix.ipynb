## 1.3. Analyser et critiquer les équations de Lotka et Volterra

Dans l'excellent MOOC [«Python pour les scientifiques»](https://courses.ionisx.com/courses/EPITA/002-001-002/2014-FALL/course/)  d'Olivier Ricou (EPITA), j'ai trouvé ce Notebook, que j'ai modifié pour vous. Il présente le modèle proies-prédateurs ainsi :

Les équations de <a href="http://en.wikipedia.org/wiki/Lotka%E2%80%93Volterra_equation">Lotka-Volterra</a> permette de simuler l'évolution de populations
de proies et de prédateurs. Elles sont :
$$\frac{dH}{dt} = (\alpha * H - \beta \; H*P)$$

$$\frac{dP}{dt} = - P(\gamma - \delta \; H)$$

Où $H$ représente le nombre de proies herbivores et $P$ celui des prédateurs. On note que :

   * $\alpha$ représente la capacité de reprodution des proies
   * $\beta$ représente l'appétit des prédateurs
   * $\gamma$ représente le taux de morts de faim chez les prédateurs
   * $\delta$ représente le taux de reproduction des prédateurs bien nourris

### 2.1. Tracer les courbes qui montrent l'evolution du nombre de proies et de prédateurs dans le temps

Pour tracer les courbes des effectifs des proies et des prédateurs en fonction du temps, on ne va pas résoudre mathématiquement ce système de deux équations à deux inconnues, mais découper le temps en moments élémentaires. À chaque instant, on peut calculer $\frac{dH}{dt}$  et  $\frac{dP}{dt}$  et les intégrer. 
la fonction f renvoie deux valeurs qui sont $\frac{dP}{dt}$ et $\frac{dH}{dt}$. 
puis on intègre cette fonction (avec `odeint`) en lui donnant le temps défini par t et H(0) et P(0), c'est à dire les populations de proies et de prédateurs initiales au temps 0.

```python
from scipy.integrate import odeint #importer la fonction intégrale
from numpy import linspace, arange #importer la fonction linspace pour découper le temps en moments élémentaires
#import numpy as np
#from matplotlib.text import OffsetFrom

def f(Y,t,a,b,c,d):
    return Y[0]*(a-b*Y[1]), -Y[1]*(c-d*Y[0]) # attention à la porté des variables, a b c d doivent être des variables globales
#le paramètre Y est utilisé ainsi : Y(0)=H(t) et Y(1)=P(t)
#la fonction f renvoie dH/dt , dP/dt à l'instant t, pour les paramètres a b c d choisis.

t = linspace(0,10,1000) #débuter t à 0, finir à 10, en faisant 1000 pas
res = odeint(f, (10,4), t, args=(a,b,c,d)) #res renvoie la valeur de l'intégrale de f avec H(0)=10 et P(0)=4. 
# (10=10 000 lièvres et 4=4000 lynx au démarrage, on a un facteur 1000)
# pour le temps t et en utilisant les arguments a b c d défini globalement dans la cellule précédente.

figure(figsize=(10,3), dpi=100) # taille en inchs et résolution de la figure à tracer
plot(t,res) #trace le graphique 

#Annotation du graphique
import numpy as np
import matplotlib.pyplot as plt #importer la bibliothèque plot pour légender les axes
#plt.legend(loc='best', fontsize=8) #au meilleur emplacement
plt.title("Évolution de la population de proies et de prédateurs, selon Lotka et Volterra", fontsize=12)
plt.xlabel("Temps en multiple de 5 ans", fontsize=10) #titre des abscisses
plt.ylabel("Effectifs des populations en milliers", fontsize=10)
plt.text(-0.4,10,"Proies", color='b', fontsize=12, fontfamily="arial", bbox=dict(facecolor='blue', alpha=0.1)) 
#position x,y, texte, couleur, taille, police, rectangle transparent en fond
plt.text(-0.4,0,"Prédateurs", color='orange', fontsize=12, fontfamily="arial", bbox=dict(facecolor='orange', alpha=0.1))
#si vous voulez changer la longueur de l'axe des abscisses, ordonnées plt.axis([0,10,0,25])
grid() #par défaut montre un quadrillage gris. grid(b=None) supprime le quadrillage
```

Pour tracer ce cycle, regardons la variable `res` afin de comprendre sa structure :

Il s'agit d'un tableau donnant H(t) en 1ère position et P(t) en 2ème, pour différentes valeurs de t.
On peut donc dessiner la seconde colonne P(t) en fonction de la première H(t), soit `res[:,1]` en fonction de `res[:,0]` :

```python
t = linspace(0,3,100)           # pas trop de cycles 
init_pred= 3   # effectif de predateurs pour 10 proies, en milliers
a = 3   # réattribué au cas ou l'utilisateur les aurait modifié précédemment.
b = 2  
c = 3   
d = 0.3
res = odeint(f, (10,init_pred), t, args=(a,b,c,d)) 

plot(res[:,0],res[:,1],'.',label="%2.1f"%(init_pred)) # trace H(t) en fonction de P(t).
#label met une légende qui donne l'effectif de predateurs pour 10 proies avec deux chiffes avant la virgule et 1 après.
# '.' demande de mettre un point par valeur. 
xlabel(u'Effectifs des proies en milliers')
ylabel(u'Effectifs des prédateurs en milliers')
plt.title("Effectifs des proies en fonction des effectifs de prédateurs, selon Lotka et Volterra", fontsize=14)
legend()
```

```python
t = linspace(0,3,100)           # pas trop de cycles 
for init_pred in range(1,7):   # on fait varier effectif initial de predateurs pour 10 proies entre 1 et 7
  res = odeint(f, (10,init_pred), t, args=(a,b,c,d)) 
  #res renvoie la valeur de l'intégrale de f avec H(0)=10 et P(0)=init_pred. 
  plot(res[:,0],res[:,1],'.',label="%2.1f"%(init_pred))  
xlabel(u'Effectifs des proies en milliers')
ylabel(u'Effectifs des prédateurs en milliers')
plt.title("Effectifs des proies en fonction des effectifs de prédateurs, selon Lotka et Volterra", fontsize=14)
legend()
```

### 2.4. Trouver la valeur d'équilibre qui fait que les deux populations restent constantes

__Surprise__ La courbe correspondant à 2 prédateurs pour 10 proies est incluse dans la courbe 1.0 alors qu'ensuite chaque courbe est incluse dans celle de ratio plus petit !

Zoomons en choisissant des ratios proie/prédateur dans la zone d'inversion :

```python
t = linspace(0,2.5,100)   # temps pas assez long pour avoir des cycles complets
for init_pred in arange(0.5,3,0.5):   # arange pour avoir des réels, on avance d'un pas de 0.5
  res = odeint(f, (10,init_pred), t, args=(a,b,c,d)) 
  plot(res[:,0],res[:,1],'.',label="%2.1f"% (init_pred)) 
xlabel(u'proies')
ylabel(u'prédateurs')
legend()
grid()  # on commence les cycles en x = 10
```

Quand il y a 1,5 prédateurs pour 10 proies (avec les coefficients choisis), aucune des deux populations ne fluctue plus !

### Partie facultative

Calculons la ratio qui mène au cercle le plus petit c.a.d à l'équilibre naturel. On note qu'il s'agit du cycle le plus court, on va donc cherche à minimiser le cycle et pour cela on va utiliser `minimize` de `scipy.optimize`.

Pour commencer définissons la fonction à optimiser :

```python
def ecart(ratio):

    t = linspace(0,9,1000)
    res = odeint(f, (10., 10./ratio), t, args=(a,b,c,d))
    return max(res[:,0]) - min(res[:,0])

from scipy.optimize import minimize  # attention, demande la version 0.11 ou + de Scipy

mini = minimize(ecart,[3],method='TNC',bounds=[(1,10)])
print ("Minimum pour le ratio proies/prédateurs= %f" % mini.x[0])
t = arange(0,5,0.01)
res = odeint(f, (10., 10./mini.x[0]), t, args=(a,b,c,d))
#plot(res[:,0],res[:,1])
```

Le ratio de 20 proies /3 prédateurs donne la position d'équilibre. 

```python
plot(t,res) #graphique pour le ratio à l'équilibre
```
